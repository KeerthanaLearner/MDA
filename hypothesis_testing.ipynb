```python
import numpy as np
from scipy.stats import f,chi2
```

```python
def compute_T_square(df,mu_0):
  n=len(df)
  meanVec=np.mean(df,axis=0)
  covMat=np.cov(df,rowvar=False)
  value=n*(meanVec-mu_0).T.dot(np.linalg.inv(covMat)).dot(meanVec-mu_0)
  return value

def compute_tabulated_value(df,alpha):
  #based on the number of observations decide the distributopn of T^2 and
  n=len(df)
  p=len(df[0])
  if n-p>40:
    #large sample
    tab_value = chi2.ppf(1-alpha,p)
  else:
    tab_value=f.ppf(1-alpha,p,n-p)*((n-1)*p/(n-p))
  return tab_value

def make_decision(computed_T_square,table_value):
  if computed_T_square>=table_value:
    result=print("Reject the null hypothesis")
  else:
    result=print("Fail to reject the null hypothesis")
  return result
```

```python
mu_0=np.array([7,11])
df=np.array([[2,12],[8,9],[6,9],[8,10]])
T_square=compute_T_square(df,mu_0)
T_square
```

```python
table_value=compute_tabulated_value(df,0.05)
table_value
```

```python
make_decision(T_square,table_value)
```

hence we reject H0

**confidence interval**

```python
def compute_confidence_intervals(df, meanVec, covMat, alpha):
    n = len(df)
    p = len(df[0])
    # Critical value from F-distribution
    critical_value = f.ppf(1 - alpha, p, n - p) * (p * (n - 1) / (n - p))

    # Compute confidence intervals for each mean
    conf_intervals = []
    for i in range(p):
        margin_of_error = np.sqrt(critical_value * covMat[i, i] / n)
        ci_lower = meanVec[i] - margin_of_error
        ci_upper = meanVec[i] + margin_of_error
        conf_intervals.append((ci_lower, ci_upper))

    return conf_intervals
```

```python
# Compute confidence intervals
meanVec=np.mean(df,axis=0)
covMat=np.cov(df,rowvar=False)
conf_intervals = compute_confidence_intervals(df, meanVec, covMat, 0.05)
print("Confidence Intervals:", conf_intervals)
```
